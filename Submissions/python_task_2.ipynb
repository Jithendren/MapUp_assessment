## 3.Finding IDs within percentage Threshold

```python
def find_ids_within_ten_percentage_threshold(dataset):
    data3=pd.read_csv(dataset)
```

```python
data3
```

```python
average_distance=data3['distance'].mean()
```

```python
average_distance
```

```python
# Calculate the threshold values for 10% above and below the average
lower_threshold = average_distance - (average_distance * 0.1)
upper_threshold = average_distance + (average_distance * 0.1)
```

```python
# Filter rows within the 10% threshold
within_threshold = data3[(data3['distance'] >= lower_threshold) & (data3['distance'] <= upper_threshold)]
```

```python
# Get unique values from 'id_start' column and sort them
result_list = sorted(within_threshold['id_start'].unique().tolist())
```

## 4.Calculate Toll Rate

```python
def calculate_toll_rate(dataset):
    data3=pd.read_csv(dataset)
```

```python
rate_coefficients = {'moto': 0.8, 'car': 1.2, 'rv': 1.5, 'bus': 2.2, 'truck': 3.6}
```

```python
# Calculate toll rates for each vehicle type
for vehicle_type, rate_coefficient in rate_coefficients.items():
    column_name = f'{vehicle_type}_rate'
    data3[column_name] = data3['distance'] * rate_coefficient
```
